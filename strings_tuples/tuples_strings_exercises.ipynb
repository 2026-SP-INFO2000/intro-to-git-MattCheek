### Exercise 3
Check if a string is the same forward and backward (Palindrome). Use the input function to input a string.

```python
string = input("Enter the string you want to check: ")
list = []
for x in string:
    list.append(x)
if list == list[::-1]:
    print(f'{string} is a palindrome.')
else:
    print(f'{string} is not a palindrome.')
```

### Exercise 4 Password Strength Checker 
Check if a password has at least 1 digit, 1 uppercase, 1 lowercase letter, and 1 special character from `!@#$%^&*`

```python
pwd = "PyTh0n123"  # or you can use the input function to input your own passwords
charList = ["!","@","#","$","%","^","&","*"]
counter = 0
if pwd == pwd.upper():
    print("Password does not contain at least 1 lowercase letter")
if pwd == pwd.lower():
    print("Password does not contain at least 1 uppercase letter")
if len(pwd) < 1:
    print("Password does not contain at least 1 digit")
for char in pwd:
    if char not in charList:
        counter +=1
if counter == len(pwd):
    print("Password does not contain at least 1 special character")
else:
    print("Your password is secure!")
```
