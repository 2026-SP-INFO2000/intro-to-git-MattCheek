### Exercise 1: number guessing game

Redo using the for loop yourself.

Given a target value, input your guess, and determine the result of your guess based on the following conditions
- If guess ==target, print "Correct!"
- If |guess-target|=1, print "Close!"
- If guess > target+1, print "Too High"
- If guess < target -1, print "Too Low"

```python
import random
num = random.randint(0,10)
guesses = 10
for i in range(guesses):
    guess = int(input("Input your guess: "))
    if guess == num:
        print("You won!")
        break
    elif abs(guess-num) == 1:
        print("Close!")
    elif guess > num+1:
        print("Too high")
    elif guess < num-1:
        print("Too low")
```

### Exercise 2

Given `names = ["Alice", "Bob", "Charlie", "Diana"]`, print greetings for names with more than 3 letters (e.g., `Hello Alice!`).

```python
names = ["Alice", "Bob", "Charlie", "Diana"]
for name in names:
    if len(name)>3:
        print(f'Hello, {name}.')
```

### Exercise 3 — Average Score 
Given `scores = [80, 95, 67, 88, 92]`, compute and print the average of the scores that are higher than 85.

```python
scores = [80, 95, 67, 88, 92]
highScores = []
for score in scores:
    if score > 85:
        highScores.append(score)
sum = 0
for score in highScores:
    sum = sum+score
print(f'The average of the scores greater than 85 is {round(sum/len(highScores),2)}')

    
```


### Exercise 4 
Write the program to print the Pascal Triangle **upside-down**.

```python
rows = 6   
triangle = []
for i in range(rows):
    row = [1] * (i + 1)
    for j in range(1, i):
        row[j] = triangle[i-1][j-1] + triangle[i-1][j]
    triangle.append(row)
for row in reversed(triangle):
    print(row)
```

### Exercise 5
Generate a list that contains pairs `(x, y)` where `x` is in `[1, 2, 3, 4]`, `y` is in `[1, 2, 3, 4]`, and the product of `x` and `y` is even. Use both nested for loops and list comprehension.

```python
pairs = []
for x in [1, 2, 3, 4]:
    for y in [1, 2, 3, 4]:
        if (x * y) % 2 == 0:
            pairs.append(x * y)
print(pairs)
```

```python
pairs = [x*y for x in [1,2,3,4] for y in [1,2,3,4] if (x*y) % 2==0]
print(pairs)
```
