## Exercise 3
The Body Mass Index or BMI is calculated using the following formula: 
$\textbf{BMI} = \frac{\text{Weight in Kilograms}}{\text{Height in Meters} \times \text{Height in Meters}} $
Write a program that asks the user for their height in inches, and weight in pounds and prints out their BMI. Note, you will need to convert the user inputs to metric units.

```python
height = float(input("Please enter your height in inches: "))
weight = float(input("Please enter your weight in pounds: "))
weightKG = weight/2.2
heightM = height * 0.0254
BMI = weightKG/(heightM*heightM)
print(f"Your BMI is: {BMI}")
```

## Exercise 4
Write a program that asks the user for the coefficients (a,b, and c) of a quadratic equation and computes and prints out the root (x) using the following formula:
$ ax^2+bx+c=0 \implies x=\frac{-b \pm \sqrt{b^2-4ac}}{2a}$

```python
import math
a = float(input("Please enter a: "))
b = float(input("Please enter b: "))
c = float(input("Please enter c: "))

num1 = -1*b + math.sqrt((b**2) - 4*a*c)
num2 = -1*b - math.sqrt((b**2) - 4*a*c)
denom = 2*a

root1 = num1/denom
root2 = num2/denom

print(f"Your roots are {root1} and {root2}")

```
