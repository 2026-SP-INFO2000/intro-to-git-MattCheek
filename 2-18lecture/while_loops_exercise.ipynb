### Exercise 1
Keep asking for a password until the user enters `'python123'` or the maximum number of attempts is reached (You can use the input function to input the maximum number of attempts).
If correct, print `'Access granted!'`. 

```python
attempts = int(input("Input the number of attempts: "))
while attempts != 0:
    guess = input("Please enter the password: ")
    if guess == "python123":
        print("Access granted!")
        break
    else:
        attempts -= 1
        print(f'Password incorrect! You have {attempts} attempts left')
```

### Exercise 2 Write the number guessing game using while loops. 
Keep guessing until correct or the number of guesses is exhausted.
- If guess ==target, print "Correct!"
- If |guess-target|=1, print "Close!"
- If guess > target+1, print "Too High"
- If guess < target -1, print "Too Low"


```python
import random
num = random.randint(0,10)
guess = 11
while guess != num:
    guess = int(input("Input your guess: "))
    if guess == num:
        print("Correct!")
        break
    elif abs(guess-num) == 1:
        print("Close")
    elif guess > num+1:
        print("Too high!")
    elif guess < num-1:
        print("Too low!")
```

### Exercise 3 Cookie Monster Game

Start with a number of cookies (use input function). Each loop, the “monster” eats 1–3 cookies (random). Keep looping until no cookies remain. Print how many bites it took.
- `import random` and `cookies_per_bite = random.randint(1, 3)`

```python
import random
num = int(input("How many cookies are there? "))
counter = 0
while num != 0:
    cpb = random.randint(1,3)
    num = num-cpb
    print(cpb)
    counter += 1
print(f'It took {counter} bites to finish the cookies')
```

### Exercise 4
Design your own ATM menu, with basic operations `'balance'`, `'deposit'`, `'withdraw'`, `'Exit'`.

```python
print("Welcome to the ATM")
bal = int(input("Input your balance: "))
while True:
    action = input("What would you like to do? (balance, deposit, withdraw, exit) ")
    match action:
        case "balance":
            print(f'Your balance is {bal}')
            continue
        case "deposit":
            dAmt = int(input("How much would you like to deposit? "))
            bal += dAmt
        case "withdraw":
            wAmt = int(input("How much would you like to withdraw? "))
            if wAmt > bal:
                print("Withdraw amount greater than balance. Please see clerk to take out a loan")
            else: 
                bal -= wAmt
        case "exit":
            print("Thank you for using the ATM.")
            break
        case _:
            print("Please enter a valid operation")
            
```
