# Exception Handling in Python

**Learning goals:**
- Understand what exceptions are
- Distinguish between syntax errors and runtime errors
- Use `try`, `except`, `finally`, and `else`

## 1) SyntaxError & Exception
There are (at least) two distinguishable kinds of errors: syntax errors and exceptions
- SyntaxError: caught before running
- Exception (runtime error): Errors detected during execution are called exceptions and are not unconditionally fatal


````python
# SyntaxError: caught before running
# print("Hello"   # Uncomment to see SyntaxError

# Runtime error (exception)
# Uncomment next line to see ZeroDivisionError
# x = 1 / 0
````

 **Python Built-in Errors**

| **Category**        | **Error Types** |
|----------------------|------------------|
| **Syntax & Indentation** | `SyntaxError`, `IndentationError`, `TabError` |
| **Arithmetic**       | `ZeroDivisionError`, `OverflowError`, `FloatingPointError` |
| **Type & Value**     | `TypeError`, `ValueError` |
| **Name & Attribute** | `NameError`, `UnboundLocalError`, `AttributeError` |
| **Import & Module**  | `ImportError`, `ModuleNotFoundError` |
| **Index & Key**      | `IndexError`, `KeyError` |
| **File & I/O**       | `FileNotFoundError`, `PermissionError`, `IsADirectoryError`, `EOFError` |
| **OS & System**      | `OSError`, `BlockingIOError`, `ChildProcessError`, `TimeoutError` |
| **Runtime**          | `RuntimeError`, `NotImplementedError`, `RecursionError` |
| **Assertion**        | `AssertionError` |
| **Custom**           | User-defined exceptions (subclass of `Exception`) |


**Why exception handling**
- Programs break often (wrong input, missing files, division by zero, internet errors). Without exception handling, programs just crash.

- Exception handling gives friendly user experience: Instead of a scary traceback, you can say:"Sorry, file not found. Please check the filename."

- Debugging is easier when exceptions are caught where they happen.

## 2) Basic `try/except` syntax for exception handling
```python
try:
    # code that might fail
except SomeError:
    # what to do if it fails
```
- Without exception handling → crash.
- With exception handling → program survives, user gets a clear message.

````python
try:
    a = int(input("Enter first number: "))
    print('The value of a is:', a)
except:
    print('Please enter a valid number.')
````

````python
try:
    num = int(input("Enter a number: "))
    print(10 / num)
except ZeroDivisionError:
    print("❌ You cannot divide by zero!")
except ValueError:
    print("❌ Please enter a valid integer.")
````

## 3) finally and else
```python
try:
    # codes that might fail
except SomeError:
    # what to do if it fails
else:
    # codes that excute without exception
finally:
    #codes that execute always
```

````python
try:
    f = open("data.txt", "r")
    content = f.read()
except FileNotFoundError:
    print("File not found!")
else:
    print("File read successfully.")
finally:
    print("Done (whether error or not).")
````

## 4) Raising exceptions manually
- Python raises exceptions automatically when errors occur (`ZeroDivisionError`, `ValueError`, etc.)
- But you can also manually raise exceptions with the `raise` keyword; this is useful when you want to enforce rules or signal problems in your own code
- Syntax: `raise ERRORTYPE(Message)`
- Combine with `try/except` for exception handling

````python
def withdraw(amount):
    if amount < 0:
        raise ValueError("Amount must be positive")
    print("Withdrew", amount)

withdraw(-5)  
````

````python
def withdraw(amount):
    if amount < 0:
        raise ValueError("Amount must be positive")
    print("Withdrew", amount)

try:   
    withdraw(-5) 
except ValueError:
    print("Withdraw amount cannot be negative") 
````

## 5) Why not use `if` to do error handling?
- We can use `if` to check conditions that will yield errors; however, you need to think of all the possible conditions; this works well for expected cases.
- With `try/except`, you let Python try the operation, and if something goes wrong, you catch the exception (post-hoc); Wwrks well for unexpected or multiple types of errors.
- Rule of thumb:
  - Use `if` when you can easily predict and test for a problem **before** it happens.
  - Use `try/except` when the error may come from deeper inside Python (file access, type conversion, user mistakes) and you want a safety net.

````python
filename = "data.txt"
if os.path.exists(filename):
    f = open(filename)
    data = f.read()
else:
    print("File not found!")
````

## Example 1 —  Age Checker
Write a function `check_age(age)` that:
- Raises a `ValueError` if age is less than 0 or greater than 120.
- therwise, returns "Valid age".

In the main program, ask the user for an age and handle exceptions with a friendly message.

````python
def check_age(age):
    if age < 0 or age > 120:
        raise ValueError("Age must be between 0 and 120.")
    return "Valid age"

try:
    age = int(input("Enter age: "))
    print(check_age(age))
except ValueError as e:
    print("❌ Error:", e)
````

## Example 2 — User Input Checker
Ask the user for an integer.
- If valid, print the square.
- If invalid, print `'Not an integer!'`.

````python
def squared_from_input(text):
    try:
        n = int(text)
        return n*n
    except ValueError:
        return "Not an integer!"

# Demo without interactive input
print(squared_from_input("9"))   # 81
print(squared_from_input("abc")) # Not an integer!
````

## Example 3 — Shopping Cart Validator 🛒
Define a function `checkout(cart)` where `cart` is a dictionary of items → prices.
- If any price is negative, raise `ValueError`.
- Otherwise, return the total.

````python
def checkout(cart):
    total = 0.0
    for item, price in cart.items():
        if price < 0:
            raise ValueError(f"Negative price for {item}: {price}")
        total += price
    return total

print(checkout({"apple": 1.0, "banana": 0.5}))
try:
    print(checkout({"apple": 1.0, "bad": -2}))
except ValueError as e:
    print("Caught:", e)
````
