## Exercise 1 – Leap Year Checker
Determine if a given year is a leap year.
- A year is a leap year if divisible by 4 but not 100, unless also divisible by 400.

```python
year = int(input("What year do you want to check? "))
if year%4 == 0 and year%100 != 0:
    print("This year is a leap year.")
elif year%400 == 0:
    print("This year is a leap year.")
else:
    print("This year is not a leap year.")
    
```

## Exercise 2– ATM Withdrawal
Given a balance in your bank account, input an withdrawal amount, check if it is a valid withdrawal, and return the remaining balance for a valid withdrawal

```python
balance = 100
wdraw = int(input("Enter the amount you want to withdraw: "))
if wdraw > balance:
    print("The amount entered is greater than the balance in the account. Cannot complete withdraw.")
else:
    print(f'Withdrew {wdraw} dollars. Remaining balance: {balance-wdraw}')
```

## Exercise 3– Number Guessing Game
Given a target value. Run the following code cell to randomly generate a value between 1 and 10.

```python
import random

target = random.randint(1, 10)   # random int between 1 and 10 (inclusive)
```

Input your guess, and determine the result of your guess based on the following conditions
- If guess ==target, print "Correct!"
- If |guess-target|=1, print "Close!"
- If guess > target+1, print "Too High"
- If guess < target -1, print "Too Low"

Run your program in the following code cell multiple times until you find the target value. You can use a counter to track how many guesses you make.

```python
counter = 0
```

```python
counter +=1
guess = int(input("Enter your guess: "))
if guess == target:
    print(f"Correct! Guess number: {counter}")
elif abs(guess-target) == 1:
    print("Close! Try again.")
elif guess > target+1:
    print("Too high.")
elif guess < target-1:
    print("Too low.")
```

## Exercise 4 – Grading with Plus/Minus
 Input a score and return the corresponding grade based on the following 
 - A+: 97-100
 - A: 93-96
 - A-: 90-92
 - B+: 87-89
 - B: 83-86
 - B-: 80-82
 - C+: 77-79
 - C: 73-76
 - C-: 70-72
 - D: 60-69
 - F: <60 

```python
score = int(input("Enter the score: "))
if score >= 97:
    print("A+")
elif score >= 93: 
    print("A")
elif score >= 90: 
    print("A-")
elif score >= 87: 
    print("B+")
elif score >= 83: 
    print("B")
elif score >= 80: 
    print("B-")
elif score >= 77: 
    print("C+")
elif score >= 73:
    print("C")
elif score >= 70: 
    print("C-")
elif score >= 60: 
    print("D")
elif score < 60:
    print("F")
```

### Exercise 5
Write a program that asks for a user's exam score and whether they submitted all assignments (yes/no). Print *Pass* if score ≥ 60 and all assignments were submitted, otherwise print *Fail*. Try solving with both a single condition and nested conditionals.

```python
score = int(input("Enter your exam score: "))
subStatus = input("Have you submitted all assignments? (y/n): ")

if score >= 60 and subStatus.lower() == "y":
    print("Pass")
else:
    print("Fail")
    
    
```
