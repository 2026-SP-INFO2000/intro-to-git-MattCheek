1. Given `text = "programming is fun"`, use a dictionary to show how many times each of vowels `(a, e, i, o, u)` appear in the text, print the dictionary. (5 points)

````python
vowels = ['a', 'e', 'i', 'o', 'u']
text = "programming is fun"
voweldict = {'a': 0, 'e': 0, 'i': 0, 'o': 0, 'u': 0}
for char in text:
    if char in voweldict:
        voweldict[char] += 1
print(voweldict)
````

2. You have a dictionary representing a store’s inventory:
```python
inventory = {"apple": 5, "banana": 2, "orange": 0}
```
- Ask the user what they want to buy.
- If the item is available (quantity > 0), reduce the count by 1 and confirm purchase.
- If not, print `'Sorry, out of stock'`.
- Stop when the user types `'quit'`.  (5 points)

````python
inventory = {'apple': 5, 'banana': 2, 'orange': 0}
while True:
    item = input("What would you like to purchase ('quit' to quit)? ")
    if item == "quit":
        break
    elif inventory[item] > 0:
        inventory[item] -= 1
        print("Purchase confirmed!")
    elif inventory[item] == 0:
        print("Sorry, out of stock.")
    elif item not in inventory:
        print("We do not carry that item.")
````

4. Password Checker
- Write a function `check_password(pwd)`to check if a password is valid. A valid password should have least 8 characters, and has at least 1 digit, 1 uppercase, 1 lowercase letter, and 1 special character from `!@#$%^&*`.

- In the main program, use the input function to input the password and call the function `check_password(password)` to check the password. (5 points)

````python
def check_password(pwd):
    chars = ["!","@","#","$","%","^","&","*"]
    nums = ['1','2','3','4','5','6','7','8','9','0',]
    counter1 = 0
    counter2 = 0
    for char in pwd:
        if char in chars:
            counter1 += 1
    for char in pwd:
        if char in nums:
            counter2 += 1
    if counter2 == 0:
        print("Password has no digits.")
    elif pwd.lower() == pwd:
        print("No uppercase letters.")
    elif pwd.upper() == pwd:
        print("No lowercase letters.")
    elif counter1 == 0:
        print("No special characters")
    else:
        print("Password is secure!")

password = input("Please choose a password: ")
check_password(password)
    
````

5. Given the maximum number of guesses and a target value (randomly generated between 1 and 10), use for loop or while loop to simulate the number guessing game. In each loop, input your guess and determine the result of your guess based on the following conditions
- If guess ==target, print "Correct!"
- If |guess-target|=1, print "Close!"
- If guess > target+1, print "Too High"
- If guess < target -1, print "Too Low" 
- Use formatted string to print how many guesses remaining at the end of each loop. (5points)

````python
import random
guesses = 5
target = random.randint(1,10)
while guesses != 0:
    guess = int(input("Guess a number between 1 and 10"))
    if guess == target:
        print("Correct!")
        break
    elif abs(guess-target) == 1:
        guesses -= 1
        print(f'Close! You have {guesses} guess(es) left.')
    elif guess > target+1:
        guesses -= 1
        print(f'Too high! You have {guesses} guess(es) left.')
    elif guess < target-1:
        guesses -=1
        print(f'Too low! You have {guesses} guess(es) left.')
        
        
                

````

6. Write a program to print Pascal triangle **upside down** for a given number of rows. (6 points)

 A Pascal triangle is
```
Row 0:    1
Row 1:    1 1
Row 2:    1 2 1
Row 3:    1 3 3 1
Row 4:    1 4 6 4 1
```

````python
rows = 5  
triangle = []
for i in range(rows):
    row = [1]*(i+1)
    for j in range(1,i):
        row[j] = triangle[i-1][j-1] + triangle[i-1][j]
    triangle.append(row)
    
triangle.reverse()
for row in triangle:
    print(row)

    
    
    

    
````
