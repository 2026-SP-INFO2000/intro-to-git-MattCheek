## Exercise 1 — Joke Writer 🤡
- Create a file called `jokes.txt`;

- Write 3 jokes into it (each on a new line);

- Then reopen it, read all lines, and print them with line numbers.

```python
# Expected output example:
# 1: Why did the Python cross the road? To get to the other IDE!
# 2: ...
# 3: ...
jokes = """A horse walks into a bar, and the bartender goes: 'Why the long face?'
What does an apple work out the most? The core!
What did the number 5 say to the square root of negative one? 'Get real'"""
with open("jokes.txt", "w") as f:
    f.write(jokes)
with open("jokes.txt", "r") as f:
    counter = 1
    for line in f:
        print(f'{counter}: {line.rstrip()}')
        counter +=1
    
    
```
