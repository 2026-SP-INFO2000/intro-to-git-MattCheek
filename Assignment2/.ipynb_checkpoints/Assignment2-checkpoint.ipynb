```python
while True:
    ans = input("What is the answer to the Great Question of Life, the Universe and Everything? (q to quit): ")
    match ans.lower():
        case "42":
            print("Yup")
        case "fourty-two":
            print("Yup")
        case "fourty two":
            print("Yup")
        case "q":
            print("Thanks for trying.")
            break
        case _:
            print("No")
        
        
```

Problem 3

```python
users = ["Matt", "Charlie", "Laura", "Ben"]
passwords = ["123", "456", "abc", "def"]
grades = [["Math: B", "Physics: C", "Geography: A", "Chemistry: B"], 
          ["Math: C", "Physics: C", "Geography: B", "Chemistry: D"],
          ["Math: A", "Physics: B", "Geography: B", "Chemistry: A"],
          ["Math: A", "Physics: A", "Geography: A", "Chemistry: B"]]

while True:
    uCheck = input("Welcome to the database! Please enter your name: ")
    if uCheck in users:
        pCheck = input("Please enter your password: ")
        if pCheck in passwords:
            if users.index(uCheck) != passwords.index(pCheck):
                print("Incorrect password. Try again.")
            elif users.index(uCheck) == passwords.index(pCheck):
                print("Access granted. Displaying grades.")
                print(grades[users.index(uCheck)])
        elif pCheck not in passwords:
            print("Incorrect password. Try again.")
                
    elif uCheck not in users:
        nUser = input("Name not in database. Please enter your name to create a new account: ")
        nPassword = input("Please create a passowrd.")
        
        nGrades = []
        Math = input("Please enter your letter grade for Math: ")
        nGrades.append("Math: " + Math)
        Physics = input("Please enter your letter grade for Physics: ")
        nGrades.append("Physiscs: " + Physics)
        Geography = input("Please enter your letter grade for Geography: ")
        nGrades.append("Geography: " + Geography)
        Chemistry = input("Please enter your letter grade for Chemistry: ")
        nGrades.append("Chemistry: " + Chemistry)
        
        grades.append(nGrades)
        passwords.append(nPassword)
        users.append(nUser)
        
    
```

Problem 4

```python
import random
fruits =  ["apple", "banana", "orange", "grape",
"kiwi", "mango", "strawberry", "pineapple", "watermelon", "cherry"]
while True:
    name = input("Hello user! Please enter your name (q or Q to quit): ")
    if name.lower() == "q":
        break
    else:
        num = random.randint(0,10)
        index = random.randint(0,9)
        if num == 0:
            print(f'Sorry {name}, no fruit for you today.')
        elif num == 1:
            print(f'Hello {name}, you get {num} {fruits[index]}. Enjoy!')
        else: 
            print(f'Hello {name}, you get {num} {fruits[index]}s. Enjoy!')
```

Problem 5

```python
import random
import string
length = int(input("How many characters should your password be? "))
chars = ['!', '@', '#', '$','%','^','&']
pWord = []
for i in range(length//2):
    char1 = random.choice(string.ascii_letters)
    char2 = random.choice(chars)
    pWord.append(char1)
    pWord.append(char2)
password = ''.join(pWord)
print(f'Your passowrd is: {password}')
    
    
```
