Problem 3

```python
users = ["Einstein", "Jordan", "Tesla", "Curie"]
passwords = ["123", "456", "abc", "def"]
grades1 = {
"Einstein" : ["5/5", "50/50", "10/10", "10/10"],
"Jordan" : ["4/8", "50/57", "7/10", "10/18"],
"Tesla" : ["8/10", "22/25", "3/5", "5/5"],
"Curie" : ["3/3", "20/20"]
}
grades2 = {
"Hawking" : ["10/10"],
"Newton" : ["30/30"],
"Gandhi" : ["100/100"],
"Cruise" : ["1/1"]
}
grades3 = {
"Turing" : ["10/10", "2/4"],
"Ali" : ["7/9", "2/3"]
}

while True:
    uCheck = input("Welcome to the database! Please enter your name (or q to quit): ")
    if uCheck in users:
        pCheck = input("Please enter your password: ")
        if pCheck in passwords:
            if users.index(uCheck) != passwords.index(pCheck):
                print("Incorrect password. Try again.")
            elif users.index(uCheck) == passwords.index(pCheck):
                print("Access granted. Displaying grades.")
                print(grades1[uCheck])
        elif pCheck not in passwords:
            print("Incorrect password. Try again.")
    elif uCheck == "q":
        break
                
    elif uCheck not in users:
        nUser = input("Name not in database. Please enter your name to create a new account: ")
        nPassword = input("Please create a passowrd.")
        
        nGrades = []
        Math = input("Please enter your grade for Math: ")
        nGrades.append(Math)
        Physics = input("Please enter your grade for Physics: ")
        nGrades.append(Physics)
        Geography = input("Please enter your grade for Geography: ")
        nGrades.append(Geography)
        Chemistry = input("Please enter your grade for Chemistry: ")
        nGrades.append(Chemistry)
        
        grades1[nUser] = nGrades
        passwords.append(nPassword)
        users.append(nUser)

while True:
    passed = []
    listCheck = input("What grade list would you like to check? (grades1, grades2, grades3, q to quit): ")
    match listCheck:
        case "grades1":
            listCheck = grades1
        case "grades2":
            listCheck = grades2
        case "grades3":
            listCheck = grades3
        case "q":
            break
        case _:
            print("Invalid input.")
            continue
    for name, score in listCheck.items():
        counter = 0
        for char in score:
            nums = char.split("/")
            if nums[0] == nums[1]:
                counter += 1
            if counter == len(listCheck[name]):
                passed.append(name)
    print(f'Here is the list of students who passed: {set(passed)}')
                
```

```python
import random
fruits =  ["apple", "banana", "orange", "grape",
"kiwi", "mango", "strawberry", "pineapple", "watermelon", "cherry"]
while True:
    name = input("Hello user! Please enter your name (q or Q to quit): ")
    if name.lower() == "q":
        break
    else:
        num = random.randint(0,10)
        index = random.randint(0,9)
        if num == 0:
            print(f'Sorry {name}, no fruit for you today.')
        elif num == 1:
            print(f'Hello {name}, you get {num} {fruits[index]}. Enjoy!')
        else: 
            print(f'Hello {name}, you get {num} {fruits[index]}s. Enjoy!')
```

```python
import random
import string
length = int(input("How many characters should your password be? "))
chars = ['!', '@', '#', '$','%','^','&']
pWord = []
for i in range(length//2):
    char1 = random.choice(string.ascii_letters)
    char2 = random.choice(chars)
    pWord.append(char1)
    pWord.append(char2)
password = ''.join(pWord)

print(f'Your passowrd is: {password}')
    
    
```
