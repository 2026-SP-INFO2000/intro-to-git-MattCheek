### Asks for first and last name, then prints your full name along with the length of each individual name

```python
name = input("What is your name? ")
lname = input("What is your last name? ")
print(name + " " + lname)
print("The length of your first name is " + str(len(name)) + " letters.")
print("The length of your last name is " + str(len(lname)) + " letters.")
```
